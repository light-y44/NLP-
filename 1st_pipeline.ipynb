```python
import json

# Specify the path to the JSON file you want to convert
json_file_path = '/Users/devanshgupta/Desktop/work_data.json'
# Open and read the JSON file
with open(json_file_path, 'r') as json_file:
    data = json.load(json_file)
```

```python
import nltk

nltk.download('punkt')
```

```python
from nltk.tokenize import sent_tokenize
# Tockenizing by sentences.

tokenized_dict = {}
for key, value in data.items():
    value = ' '.join(value)
    sentences = sent_tokenize(value)
    tokenized_dict[key] = sentences
```

```python
#This function prints the values of the first n key. 

def print_first_few_entries(dictionary, num_entries):
    sliced_dict = {key: dictionary[key] for key in list(dictionary)[:num_entries]}
    for key, value in sliced_dict.items():
        print(f"{key}: {value}")

print_first_few_entries(data,7)
```

```python
print_first_few_entries(tokenized_dict,3)
```

```python
# This is a dictionary that stores the starting and ending index of sentences which belong to each topic/ key.

No_sent_para = {}
i = 0
for key in tokenized_dict.keys():
    No_sent_para[key] = range(i,i+len(tokenized_dict[key]))
    i = i + len(tokenized_dict[key])
```

```python
print_first_few_entries(No_sent_para,3)
```

```python
all_sentences = []
# making a list that contains all the tockenized sentences.
for sentences in tokenized_dict.values():
    print(sentences)
    all_sentences.extend(sentences)
```

```python
from sentence_transformers import SentenceTransformer

# loading a pre-trained transformer.
model_name = 'bert-base-nli-mean-tokens'
model = SentenceTransformer(model_name)
```

```python
#Embedding the sentences using S-BERT transformer. 

embedded_dict = {}
i = 0
for sentence in all_sentences:
    embedded_dict[i] = model.encode(sentence)
    if(i % 100 == 0):
        print(i)
    i = i+1
```

```python
print_first_few_entries(embedded_dict,3)
```

```python
from scipy.spatial.distance import cosine
```

```python
import time
```

```python
# using cosine similarity, to measure distance between the query and the sentences of the corpus. 

start = time.time()

context = {}
query = "Information about deadly nautral disaster."
q_embed = model.encode(query)

for i, vector in embedded_dict.items():
    similarity_score = 1 - cosine(vector, q_embed)
    context[i] = similarity_score

# Sorting the sentences based on similartiy score.
sorted_context = sorted(context.items() , key = lambda item: item[1], reverse = True)

stop = time.time()

print(stop - start)
```

```python
print(sorted_context)
```

```python
# storing the first 15 sentences in the output dict.
i = 0
output = {}
for tup in sorted_context:
    if(i == 15):
        break
    print(tup[0])
    output[tup[0]] = all_sentences[tup[0]]
    i = i+1
```

```python
# printing the sorting sentences. 
myKeys = list(output.keys())
myKeys.sort()
sorted_dict = {i: output[i] for i in myKeys}

print(sorted_dict.values())
```
